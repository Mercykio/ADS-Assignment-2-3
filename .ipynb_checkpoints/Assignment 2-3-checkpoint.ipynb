```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style("whitegrid")
```

```python
Attritiondata = pd.read_csv("WA_Fn-UseC_-HR-Employee-Attrition.csv")
```

```python
Attritiondata.head()
```

```python
Attritiondata.info()
```

# 1. Show a breakdown of distance from home by job role and attrition.

```python
df_Attrition1 = Attritiondata.groupby (["JobRole","Attrition"])["DistanceFromHome"].mean().reset_index()
df_Attrition1
```

# 2. Compare average monthly income by education and attrition.

```python
df_Attrition2 = Attritiondata.groupby (["Education","Attrition"])["MonthlyIncome"].mean().reset_index()
df_Attrition2
```

# 1. Visualization of distance from home by job role and attrition.Â¶

```python
plt.figure(figsize=(15,10))
sns.barplot(x="JobRole", y="DistanceFromHome", hue="Attrition", data=df_Attrition1)
plt.title("Employee Attrition based on Job Role and Distance From Home")
plt.xlabel("Job Role")
plt.ylabel("Distance From home")
plt.xticks(rotation=45)
```

# On average, for most of the Job roles, a longer distance from home results to a higher employee attrition

# 2. Vsualization of average monthly income by education and attrition

```python
plt.figure(figsize=(15,10))
sns.barplot(x="Education", y="MonthlyIncome", hue="Attrition", data=df_Attrition2)
plt.title("Employee Attrition based on Education and Average Monthly Income")
plt.xlabel("Education")
plt.ylabel("Average Monthly Income")
```

# A lower employee attrition is recorded at all eductaion levels for high average Monthly Income.
